```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
min_z   = 0.01
max_z   = 5
delta_z = 0.05
zbins   = np.arange(min_z,max_z+delta_z,delta_z)
zbinsc  = zbins[:-1]+(zbins[1]-zbins[0])/2.
```

```python
outpath = '/project/chihway/raulteixeira/data/'
```

```python
nz_samples = np.load(outpath+'nz_samples.npy')
```

```python
colors = ['blue', 'orange', 'green', 'red']
plt.xlim(0, 2)
plt.ylim(0, .25)
for tomobin in range(4):
    plt.plot(zbinsc, nz_samples[::100,tomobin].T, color=colors[tomobin], lw=.01)
plt.show()
```
